```python
# import packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# init url
url="http://feeds.bbci.co.uk/news/rss.xml"
```

```python
response=requests.get(url)
soup=BeautifulSoup(response.text,"html.parser")
print(soup)
```

```python
# find  all item
items=soup.findAll('item')
```

```python
# display the first item
item=items[0]
print(item.title.text)
print(item.description.text)
```

```python
news_items=[]

for i in items:
    news_i={}
    news_i['title']=i.title.text    
    news_i['description']=i.description.text
    news_i['date']=i.pubdate.text
    news_items.append(news_i)
```

```python
news_items
```

```python
# convert this list in a data base
df=pd.DataFrame(news_items,columns=['title','description','date'])
```

```python
df.head(10)
```

```python
# generate csv file
df.to_csv('web-scraping.csv',index=False,encoding='utf-8')
```
