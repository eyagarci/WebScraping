```
from selenium import webdriver
import pandas as pd
options=webdriver.ChromeOptions()
options.add_argument('-headless')
options.add_argument('-no-sandbox')
options.add_argument('-disable-dev-shm-usage')
```

```
driver=webdriver.Chrome('chromedriver',options=options)
driver.get('https://www.youtube.com/c/TensorFlow/videos')
```

```
title_channel=driver.find_elements_by_xpath('//*[@id="video-title"]')

titles=[]
for i in range(len(title_channel)):
  title=title_channel[i].text
  titles.append(title)

titles
```

```
title_channel=driver.find_elements_by_xpath('//*[@id="video-title"]')
views_channel=driver.find_elements_by_xpath('//*[@id="metadata-line"]/span[1]')
date_channel=driver.find_elements_by_xpath('//*[@id="metadata-line"]/span[2]')

videos=[]
for i in range(len(title_channel)):
  video_info={}
  video_info['title']=title_channel[i].text
  video_info['views']=views_channel[i].text
  video_info['date']=date_channel[i].text
  videos.append(video_info)

videos
```

```
df=pd.DataFrame(videos,columns=['title','views','date'])
df
```

```
df.to_csv('webSscraping+selenium.csv', index=False,encoding='utf-8')
```
